```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
data = pd.read_csv("googleplaystoremaster.csv")
```

```python
data.head()
```

```python
data.loc[10472, ['Category','Rating','Reviews','Size','Installs','Type','Price','Content Rating','Genres','Last Updated',
                  'Current Ver','Android Ver']] = ['PHOTOGRAPHY',1.9,'19','3.0M','1000','Free','0','Everyone','PHOTOGRAPHY','February 11, 2018','1.0.19','4.0 and up']
```

```python
# remove extra charactors from installs
data['Installs']=[plus for plus in data['Installs'].str.replace('+','')]
data['Installs']=[comma for comma in data['Installs'].str.replace(',','')]
```

```python
data['Installs'] = data['Installs'].astype('int') # change data type from string to int. 
```

```python
data.head()
```

```python
data.dtypes
```

```python
scale_mapper = {'Everyone':1, 'Teen':3, 'Everyone 10+':2, 'Mature 17+':3, 'Adults only 18+':4,  
       'Unrated':0}
data['Content Rating Scale'] = data['Content Rating'].replace(scale_mapper)  

# ordinal variable: the more explicit the content, the higher the number
# drop unrated. 

data = data[data['Content Rating Scale']!=0]

data.head()
```

```python
data = data[data['Content Rating Scale']!=0]
```

```python
data.head()
```

```python
plt.hist(data['Installs'],bins=10) 
# look at distribution of installs it's pretty skewed. Going to do a log transform. 
```

```python
data['log_installs'] = np.log(data['Installs']+1) # log transform. 
```

```python
cor_data = data[['Installs','log_installs','Content Rating Scale']]
```

```python
cor_data['Content Rating Scale'].unique()
```

```python
cor_data.corr()
```

```python
cor_dummy = pd.concat([data,pd.get_dummies(data['Content Rating'])], axis=1)
#Taking a look at dummy variables
```
