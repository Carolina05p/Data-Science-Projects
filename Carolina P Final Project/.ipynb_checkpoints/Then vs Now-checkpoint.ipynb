```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('covid-data.csv')
df
#Load the data
```

```python
df1 = df[['continent', 'location', 'date', 'total_cases', 'new_cases', 'total_vaccinations']]
df1
#Load the data from the colums that make sense from what I need
```

```python
df2 = df1.groupby(['continent', 'location'])['total_cases', 'new_cases', 'total_vaccinations'].min()
df2
#Data of the minimun values from the most relevants columns
```

```python
df3 = df1.groupby(['continent', 'location'])['total_cases', 'new_cases', 'total_vaccinations'].max()
df3
#Data of the maximun values from the most relevants columns
```

```python
df4 = df1.groupby(['continent', 'location'])['total_cases', 'new_cases', 'total_vaccinations'].mean()
df4
#The mean of the relevant columns needed
```

```python
df_Africa = df1[df1['continent'] == 'Africa']
df_Africa
#Data from the continent Africa
```

```python
df_Asia = df1[df1['continent'] == 'Asia']
df_Asia
#Data from the continent Asia
```

```python
df_Europe = df1[df1['continent'] == 'Europe']
df_Europe
#Data from the continent Europe
```

```python
df_Africa[["year", "month", "day"]] = df_Africa["date"].str.split("-", expand=True)
df_Africa
#Detailed data per country with on specific days in Africa
```

```python
df_Africa1 = df_Africa.groupby(['month', 'year', 'location'])['total_cases', 'total_vaccinations'].sum()
df_Africa1
#Sum of total cases and total vaccines in Africa per country
```

```python
df_North[["year", "month", "day"]] = df_North["date"].str.split("-", expand=True)
df_North
#Detailed data per country with some specific days in North America
```

```python
df_North['mon_year'] = df_North['year'] + '-' + df_North['month']
df_North
#Dataset of North America with month and year of cases
```

```python
df_North1 = df_North.groupby(['mon_year', 'location'])['total_cases', 'total_vaccinations', 'new_cases'].sum()
df_North1
#Sum of total cases, total vaccines, and new cases in North America per country
```

```python
df_North2 = df_North.groupby(['day', 'month', 'year', 'location'])['total_cases', 'total_vaccinations', 'new_cases'].sum()
df_North2
#Sum of total cases and total vaccines in North America per country with day, month and year
```
