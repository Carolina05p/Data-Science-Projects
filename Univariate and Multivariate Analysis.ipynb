```python
#From here on we start the univariate distribution plots first then we'll continue with the other plots
np.random.seed(0)
mu, sigma = 0, 1

data = np.random.normal(mu, sigma, 2500)
```

```python
p = sns.distplot(data,
                color='#F15B2A',
                hist_kws=dict(edgecolor='#F15B2A', linewidth=2),
                bins=50,
                kde=True,
                kde_kws=dict(shade=True))
```

```python
p = sns.kdeplot(data,
               color='#F15B2A',
               shade=True)
```

```python
p = sns.boxplot(data,
               palette=["#F15B2A"])
for e in p.artists:
    r, g, b, a = e.get_facecolor()
    p.set_facecolor((r, g, b, .3))
    
for i, artist in enumerate(p.artists):
    artist.set_edgecolor("#F15B2A")
    
for j in range(0,5):
    line = p.lines[j]
    line.set_color("#F15B2A")
    line.set_mfc("#F15B2A")
    line.set_mec("#F15B2A")
```

```python
p =sns.violinplot(data,
                 palette=['#F15B2A'])
```

```python
#Using matplotlib to change elements properties
parts = plt.violinplot(data,
                      vert=False,
                      showmeans=False,
                      showmedians=True)

for p in parts['bodies']:
    p.set_facecolor('#F15B2A')
    p.set_edgecolor('#F15B2A')
    p.set_alpha(0.5)
```

```python
category = ['lottery', 'buy_house', 'credit', 'party', 'sell_car', 'taxes']
data = {'amount': [12000,-65400,54632,-45000,74231,-50050]}

df = pd.DataFrame(data=data, index=category)

df.head()
```

```python
cumsum = df.amount.cumsum().shift(1).fillna(0)
df.drop(columns='cumsum', inplace=True)

df.head()
```

```python
fig = df.plot(kind='bar',
             stacked=True,
             bottom=cumsum)
```

```python
h_lines =cumsum.reset_index(drop=True).repeat(3).shift(-1)
h_lines
```

```python
h_lines[1::3]=np.nan
h_lines
```

```python
df.loc['total']=df.sum().amount
cumsum.loc['total']=0

fig =df.plot(kind='bar',
            stacked=True,
            bottom=cumsum,
            color='#F15B2A')

cumsum.loc['total'] = df[:-1].sum().amount
h_lines = cumsum.reset_index(drop=True).repeat(3).shift(-1)
h_lines[1::3]=np.nan

fig.plot(h_lines.index, h_lines.values, 'black')
```

```python
#Here starts multivariate composition plots
x = range(3, 8)
y = [[2,4,6,8,9], [10,12,14,16,5], [6,7,3,4,5]]

fig = plt.figure(figsize=(8,5))
plt.stackplot(x, y, labels=['A', 'B', 'C'], baseline='weighted_wiggle')
plt.legend(loc='upper left')
```

```python
fig = plt.figure(figsize=(8,4))

plt.bar(x, y[0], color='lightgreen', edgecolor='white', width=2)
plt.bar(x, y[1], bottom=y[0], color='skyblue', edgecolor='white', width=2)
plt.bar(x, y[2], bottom=np.add(y[0], y[1]), color='orange', edgecolor='white', width=2)
```

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
tips.loc[tips['sex']=='Male', 'size'] = -tips.loc[tips['sex']=='Male', 'size']

group_col = 'sex'
order_of_bars = tips.day.unique()[::-1]
colors = ['green', 'lightblue']
```

```python
order_of_bars
```

```python
fig = plt.figure(figsize=(10, 6))

for c, group in zip(colors, tips[group_col].unique()):
    sns.barplot(x='size', y='day', data=tips.loc[tips[group_col]==group, :],
               order=order_of_bars, color=c,
               label=group, errwidth=0)
    
plt.legend()
```
