```python
houseAge = data.houseAge
```

```python
norm = (houseAge - houseAge.mean()) / houseAge.std()
norm 
#Formula used to normalize data in tis case houseAge
```

```python
minmax = (houseAge - houseAge.min()) / (houseAge.max() - houseAge.min())
minmax
#Formula of the minimun and maximum of the data in this case houseAge
```

```python
minmax.min()
```

```python
minmax.max()
```

```python
data
```

```python
data.HouseAge = minmax
```

```python
data.houseAge = minmax
```

```python
data['houseAgeStandardized'] = norm
#Label to create a new column
```

```python
data.new_column = norm
#New column will hold the data that was just normalized in the houseAge
```
